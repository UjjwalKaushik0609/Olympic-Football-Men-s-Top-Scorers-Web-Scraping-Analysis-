```python
players = soup.find('tbody').find_all('tr', class_="ssrcss-qqhdqi-TableRowBody e1icz100")

for i, player in enumerate(players):
    try:
        # Safely extract each field
        name_tag = player.find('div', class_='ssrcss-m6ah29-PlayerName e1n8xy5b1')
        team_tag = player.find('div', class_='ssrcss-qvpga1-TeamsSummary e1n8xy5b0')
        goals_tag = player.find('div', class_='ssrcss-18ap757-CellWrapper ef9ipf0')
        stats = player.find_all('div', class_='ssrcss-1vo7v3r-CellWrapper ef9ipf0')

        # Skip if any of the main fields are missing
        if not (name_tag and team_tag and goals_tag and len(stats) >= 3):
            print(f"Skipping incomplete row at index {i}")
            continue

        player_name = name_tag.get_text(strip=True)
        team_name = team_tag.get_text(strip=True)
        goals_scored = int(goals_tag.get_text(strip=True))
        assists_made = int(stats[0].get_text(strip=True))
        matches_played = int(stats[2].get_text(strip=True))
        shots_taken = int(stats[-3].get_text(strip=True))

        # Append only if all fields are successfully extracted
        player_names.append(player_name)
        team_names.append(team_name)
        goals.append(goals_scored)
        assists.append(assists_made)
        num_matches.append(matches_played)
        shots.append(shots_taken)

    except Exception as e:
        print(f"Error at row {i}: {e}")
        continue
```

```python
print("player_names:", len(player_names))
print("team_names:", len(team_names))
print("goals:", len(goals))
```

```python
data = {
    'player_names': player_names,
    'team': team_names,
    'matches': num_matches,
    'goals': goals,
    'assists': assists,
    'shots': shots
}
df_players = pd.DataFrame(data)
```
